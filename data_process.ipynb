```python
## planning to work on the new extended dataset and merge it with our old one
#ratings based addition 
```

```python
import pandas as pd
import numpy as np
```

```python
train_extended = pd.read_csv('train_extended_post_process.csv')
```

```python
train_extended.head()
```

```python
train_extended['country'].value_counts()
```

```python
train_extended_country = train_extended[train_extended['country'].isin(['United States','Canada'])]
```

```python
train_extended_country['rating'].value_counts()
```

```python
## for usa and canade , we will have till 3.0 from 5.0 rating
train_extended_NA_rating = train_extended_country[train_extended_country['rating'].isin([3.5,3.0,4.5,5.0,4.0])]
```

```python
train_extended_NA_rating.shape
```

## now we start analysing rest of the world

```python
train_extended_world = train_extended[~train_extended['country'].isin(['United States','Canada'])]
```

```python
train_extended_world['country'].value_counts()
```

```python
train_extended_world['rating'].value_counts()
```

```python
## Rest of the world ratings are 5.0,4.5,4.0,3.5
train_extended_world_rating = train_extended_world[train_extended_world['rating'].isin([3.5,4.5,5.0,4.0])]
```

```python
train_extended_world_rating.shape
```

```python
train_extended_consolidated = pd.concat([train_extended_world_rating,train_extended_NA_rating])
```

```python
train_extended_consolidated = train_extended_consolidated.reset_index(drop = True)
```

```python
train_extended_consolidated.tail()
```

```python
## Now lets load the original processed csv to get details and merge all together
```

```python
train_filtered_v1 = pd.read_csv('train_filtered_v1.csv')
```

```python
train_filtered_v1 = train_filtered_v1.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
train_filtered_v1.reset_index(drop = True)
train_filtered_v1.head()
```

```python
train_filtered_v1.columns
```

```python
train_extended_consolidated.columns
'''missed out columns pitch, speed number_of_notes, description volume length, background_code'''
```

```python
train_filtered_v1.info()
```

```python
train_extended_consolidated['pitch'] = -1
train_extended_consolidated['speed'] = -1
train_extended_consolidated['number_of_notes'] = -1
train_extended_consolidated['description'] = -1
train_extended_consolidated['volume'] = -1
train_extended_consolidated['length'] = -1
train_extended_consolidated['background_code'] = -1
```

```python
train_filtered_v2 = pd.concat([train_extended_consolidated,train_filtered_v1])
```

```python
train_filtered_v2.shape
```

```python
train_filtered_v2.to_csv('train_filtered_v2.csv')
```
